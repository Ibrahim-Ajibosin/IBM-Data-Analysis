```python
import sqlite3 as sl

con = sl.connect('Db_Test.db')
```

```python
con.execute("""
        CREATE TABLE USER (
            id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            age INTEGER
        );
    """)
```

```python
sql = 'INSERT INTO USER (id, name, age) values(?, ?, ?)'
data = [
    (7, 'ib', 28),
    (8, 'tint', 29),
    (9, 'Mjay', 22)
]
con.executemany(sql, data)
```

```python
data = con.execute("SELECT * FROM USER")
for row in data:
    print(row)
```

```python
df = pd.read_sql('''
Select Location, date, Population, total_cases,  (total_cases/population)*100 as PercentPopulationInfected From     Covid_Death_csv order by 1,2
''', con)
df.head()
```

```python
df = pd.read_sql('''
Select Location, Population, MAX(total_cases) as HighestInfectionCount,  Max((total_cases/population))*100 as PercentPopulationInfected From Covid_Death_csv Group by Location, Population order by PercentPopulationInfected desc
''', con)
df.head()
```
